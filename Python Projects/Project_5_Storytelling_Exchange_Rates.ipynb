# Data Cleaning 

Before we dive into visualizations, it is important to clean the data

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)

exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

Notice that in the US_dollar column, the "-" value is found in the data. This is not a string

```python
#drop rows where "-" column exists and convert the US_dollar column to a float

euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar']!= '-']

euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)

euro_to_dollar['US_dollar'].value_counts()
```

# Rolling Mean

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar.head(60)
```

```python
#plot the rolling window

plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['rolling_mean'])
plt.show()
```

# How has the Euro-Dollar rate fluctuated during Covid?

For this visual, I will compare 2020 to 2016-2019 for the US to Euro Exchange Rate. As we know, 2020 was a unique year given the world faced a pandemic. 

```python
euro_to_dollar
```

```python
#Obtain list of rates only in 2020
list_2020 = (euro_to_dollar['Time'] >= '2020-01-01') &  (euro_to_dollar['Time'] <= '2020-12-31')
result_1 = euro_to_dollar[list_2020]
print(result_1)
```

```python
#Obtain list of rates only in 2016-2019

list_2016_2019 = (euro_to_dollar['Time'] >= '2016-01-01') &  (euro_to_dollar['Time'] <= '2019-12-31')
result_2 = euro_to_dollar[list_2016_2019]
print(result_2)
```

```python
#plot data on line chart 

plt.plot(result_1['Time'], result_1['US_dollar'], label = '2020')
plt.plot(result_2['Time'], result_2['US_dollar'], label = '2016-2020')
plt.legend()
plt.show()
```

From 2018 - 2019, the conversion rate from US to Euro was decreasing,

The majority of 2020 saw that exchange rate actually increase.
